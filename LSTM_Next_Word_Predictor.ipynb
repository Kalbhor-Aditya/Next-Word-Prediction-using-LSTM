```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
input_sequences=[]
for sentence in faqs.split('\n'):
  #print(sentence)
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input_sequences=pad_sequences(input_sequences,
              maxlen=max_len,
              padding='pre')
```

```python
padded_input_sequences
```

```python
X=padded_input_sequences[:,:-1]
X
```

```python
y=padded_input_sequences[:,-1]
y
```

```python
X.shape
```

```python
y.shape
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=633)
```

```python
y.shape
```

```python
y
```

## Model Building

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential([
    Embedding(633,100,input_length=126),
    LSTM(150,return_sequences=True),
    LSTM(150),
    Dense(633,activation="softmax")
])
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])
```

```python
model.summary()
```

```python
history=model.fit(X,y,epochs=100)
```

```python
import matplotlib.pyplot as plt

# Get the loss values from the history dictionary
loss = history.history['loss']

# Get the number of epochs
epochs = range(1, len(loss) + 1)

# Plot the loss values against the epochs
plt.plot(epochs, loss, label='Training loss')

# Add a title and axis labels
plt.title('Training Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# Add a legend and show the plot
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the loss values from the history dictionary
acc = history.history['accuracy']

# Get the number of epochs
epochs = range(1, len(acc) + 1)

# Plot the loss values against the epochs
plt.plot(epochs, acc, label='Training loss')

# Add a title and axis labels
plt.title('Training Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

# Add a legend and show the plot
plt.legend()
plt.show()
```

```python
import numpy as np
import time

text=input("Enter text: ")

for i in range(14):
  token_text=tokenizer.texts_to_sequences([text])[0]

  padded_token_text=pad_sequences([token_text],maxlen=126,padding='pre')

  pos=np.argmax(model.predict(padded_token_text))

  for word, index in tokenizer.word_index.items():
    if index==pos:
      text= text + " " + word
      print(text)
      time.sleep(1)
```
